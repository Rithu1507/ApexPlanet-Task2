```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Final_Cleaned_Superstore.csv")

print(df.head())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Sales"], bins=30)

plt.title("Sales Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Profit"], bins=30)

plt.title("Profit Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="Category", data=df)

plt.title("Count of Categories")

plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)
```

```python
region_profit = df.groupby("Region")["Profit"].sum()

print(region_profit)
```

```python
top_sales = df.nlargest(5, "Sales")

print(top_sales[["Category", "Sales", "Profit"]])
```

```python
avg_sales = df.groupby("Category")["Sales"].mean()

print(avg_sales)
```

```python
quantity_sold = df.groupby("Category")["Quantity"].sum()

print(quantity_sold)
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="Sales", y="Profit", data=df)

plt.title("Sales vs Profit")

plt.show()
```

```python
correlation = df[["Sales", "Profit", "Quantity", "Discount"]].corr()

print(correlation)
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(correlation, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))

region_sales.plot(kind="bar")

plt.title("Sales by Region")

plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_quantity = df["Quantity"].sum()
average_discount = df["Discount"].mean()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Total Quantity Sold:", total_quantity)
print("Average Discount:", average_discount)
```
